## Helper functions

```python
def run_on_tests(tests:list, func:function):
    passed_count = 0
    for i, test in enumerate(tests, start=1):
        inp, output = test
        print(inp, output)
        print(*inp)
        res = func(*inp)
        if output == res:
            print(f"Test {i} passed. Result '{res}' matches the Expected value '{output}'")
            passed_count += 1
        else:
            print(f"Test {i} failed. Result '{res}' doesn't match the Expected value '{output}'")
    
    print(f"Tests passed: {passed_count} / {len(tests)}. Failures: {len(tests)-passed_count}")
```

```python
class Solution:
    def removeDuplicates(self, nums: list[int]) -> int:
        l = 1

        for r in range(1, len(nums)):
            if nums[r] != nums[l - 1]:
                nums[l] = nums[r]
                l += 1
        
        return l
```
